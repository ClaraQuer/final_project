```python
reviews = reviews[reviews['month'].isin(range(2, 6))]
reviews.drop('date', axis = 1, inplace = True)
```

```python
reviews = reviews.pivot_table(index = ['month', 'year'], aggfunc = 'mean').reset_index()
reviews
```

```python
monthly_rate = month.merge(right = reviews, on = ['month', 'year'])
monthly_rate.drop(['mission_id', 'year'], axis = 1, inplace = True)
monthly_rate
```

```python
X = monthly_rate.drop('total_purchases_real', axis = 1)
y = monthly_rate['total_purchases_real']

transformer = StandardScaler().fit(X)
X_standarized = transformer.transform(X)

LR = LinearRegression().fit(X_standarized, y)
score = LR.score(X_standarized, y)
print(score)

LR.predict(X_standarized)
```
